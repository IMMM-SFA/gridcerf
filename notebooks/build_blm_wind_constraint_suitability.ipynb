# Build BLM Wind Constraint layers for GRIDCERF


The following code was used to build the BLM Wind Constraint suitability layers for GRIDCERF. GRIDCERF does not provide the source data directly due to some license restrictions related for direct redistribution of the unaltered source data.  However, the following details the provenance associated with each source dataset and how they were processed.

## 1. Setup environment

### 1.1 Download GRIDCERF

Download the GRIDCERF package if you have not yet done so from here:  https://doi.org/10.5281/zenodo.6601789.  Please extract GRIDCERF inside the `data` directory of this repository as the paths in this notebook are set to that expectation.


### 1.2 Data description


- **Title**:  West-Wide Wind Mapping Project
- **Description from Source**: The following downloads include key layers used in the West-Wide Wind Mapping Project in Esri shapefile format, which can be used in Esri ArcGIS as well as many open-source GIS applications. These files contain the same data supplied in the WWMP Data Viewer, but grouped for easier downloading.
- **Source URL**:  https://wwmp.anl.gov/maps-data/
- **Date Accessed**:  01/29/24
- **Citation**
> Argonne National Laboratory, 2016. Consolidated BLM Land with Wind Development Sensitivities. West-Wide Wind Mapping Project Maps and Data, https://wwmp.anl.gov/maps-data/
- **Application**: The West-Wide Wind Mapping Project dataset is used for identifying exclusion areas from wind siting within US BLM lands. Given that this dataset includes some of the layers that are already included in the GRIDCERF data base for wind such as national monuments, historic trails, wilderness areas, and scenic rivers, areas already included in GRIDCERF are not compiled in this notebook. The following layers are generated from this script. Layer descriptions are provided from the reference listed above. Wind siting consideration is also from the reference above.

Each of the input data files can be found in one of the following folders in the input data:
* Individual_Ecological_Resources.zip
* Individual_Potentially_Incompatible_Land_Uses.zip
* Individual_Recreation_Resources.zip
* Individual_Special_Land_Resource_Areas.zip
* Individual_Visual_Resources.zip


Most of the files below are provided as shapefiles downloadable at the link above. A select few have raster data that was provided as .adf files. These files must be converted to .tif files and saved to the source folder directory prior to running the scripts found in this notebook. This can be done in a free GIS software such as QGIS.

### 1.3 Import modules


```python
import os

import geopandas as gpd
import numpy as np
import rasterio
from rasterio import features
from rasterio.plot import show
from pathlib import Path
```

## 2. Configuration


```python
# get the parent directory path to where this notebook is currently stored
root_dir = os.path.dirname(os.getcwd())

# data directory in repository
data_dir = os.path.join(root_dir, "data")

# GRIDCERF data directory from downloaded archive
gridcerf_dir = os.path.join(data_dir, "gridcerf")

# GRIDCERF reference data directory
reference_dir = os.path.join(gridcerf_dir, "reference")

# GRIDCERF source data directory for the downloaded airport data
source_dir = os.path.join(gridcerf_dir, "source", "technology_specific",  'blm_west_wide_wind_mapping_project' )

# ecological directory
eco_dir = os.path.join(source_dir,'Individual_Ecological_Resources', 'Individual_Ecological_Resources')

# recreation directory
rec_dir = os.path.join(source_dir,'Individual_Recreation_Resources', 'Individual_Recreation_Resources')

# special land directory
special_dir = os.path.join(source_dir,'Individual_Special_Land_Resource_Areas', 'Individual_Special_Land_Resource_Areas')

# visual resources directory
vis_dir = os.path.join(source_dir,'Individual_Visual_Resources', 'Individual_Visual_Resources')

# incompatible directory
incomp_dir = os.path.join(source_dir, 'Individual_Potentially_Incompatible_Land_Uses', 'Individual_Potentially_Incompatible_Land_Uses')


# GRIDCERF technology_specific data directory
technology_specific_dir = os.path.join(gridcerf_dir, "technology_specific")

# GRIDCERF compiled final suitability data directory
compiled_dir = os.path.join(gridcerf_dir, "compiled")

# template siting mask raster
template_raster = os.path.join(reference_dir, "gridcerf_sitingmask.tif")

# template conus mask raster
land_mask_raster = os.path.join(reference_dir, "gridcerf_landmask.tif")

# template conus mask raster
conus_mask = os.path.join(reference_dir, "gridcerf_landmask.tif")

# CONUS boundary shapefile
conus_shpfile = os.path.join(reference_dir, "gridcerf_conus_boundary.shp")

# crs for converting projections
crs = 'ESRI:102003'
```

## 4. Generate suitability rasters

### 4.1 Functions to build suitability

```python
def vector_to_raster(template_raster, land_mask_raster, gdf, value_field, output_raster):
                     
    # open the template raster and extract metadata and land mask
    with rasterio.open(template_raster) as template:

        metadata = template.meta.copy()

        # update raster data type
        metadata.update(dtype=np.int16)

        # extract land mask
        land_mask_file = rasterio.open(land_mask_raster)
        land_mask = land_mask_file.read(1)
        land_mask = np.where(land_mask == 0, np.nan, 1)

        # write output raster
        with rasterio.open(output_raster, 'w+', **metadata) as out:

            out_arr = out.read(1)

            # build shapes to rasterize from target geometry and field
            shapes = ((geom, value) for geom, value in zip(gdf.geometry, gdf[value_field]))

            # burn features
            burned = features.rasterize(shapes=shapes, 
                                        fill=0, 
                                        out=out_arr, 
                                        transform=out.transform)
            
            burned = np.where(burned == 1, 1, 0).astype(np.float64)
            
            # apply land mask
            burned *= land_mask
            
            # make nan excluded
            burned = np.where(np.isnan(burned), 1, burned)

            print(output_raster)
            show(burned)

            out.write_band(1, burned.astype(np.int16))

def process_shp(input_dir, output_dir, crs, skip_list=[]):
    
    for f in os.listdir(input_dir):
        if f.endswith('.shp'):
            if f in skip_list:
                pass
            else:
                file = os.path.join(input_dir, f)
                gdf = gpd.read_file(file)
                gdf.to_crs(crs, inplace=True)
    
                # assign value for raster
                gdf['value'] = 1
    
                output_file = os.path.join(output_dir, 'gridcerf_blm_'+ Path(file).stem + '.tif')
                
                # generate raster
                vector_to_raster(template_raster=template_raster, 
                     land_mask_raster=template_raster,
                     gdf=gdf, 
                     value_field="value",
                     output_raster=output_file)
        else:
            pass
    return None

def process_raster(input_raster_file, input_dir, output_dir):

    input_raster_path = os.path.join(input_dir, input_raster_file)
    tmp_tif = os.path.join(source_dir, 'tmp.tif')

    gdal_warp_cmd = f"gdalwarp -te -2831615.228 -1539013.3223 2628318.0948 1690434.1707 -tr 1000.0 1000.0 -t_srs ESRI:102003 -dstnodata 0 -overwrite {input_raster_path} {tmp_tif}"
    os.system(gdal_warp_cmd)
    
    with rasterio.open(template_raster) as template:
        metadata = template.meta.copy()
    
        mask = template.read(1)
        
        # switch template suitability
        mask = np.where(mask == 1, 0, 1)
    
        # open tmp array
        array = rasterio.open(tmp_tif).read(1)
        array = array + mask
        array = np.where(array>=1, 1, 0)
        
        print(input_raster_file)
        show(array)
        
        output_file = os.path.join(output_dir, 'gridcerf_blm_'+ Path(input_raster_file).stem + '.tif')
        print(f'Saving raster to {output_file}')
        with rasterio.open(output_file, 'w+', **metadata) as out:
            out.write_band(1, array.astype(np.int16))

    return array
    
```

### 4.2 Generate suitability

### Ecological BLM Areas

```python
skip_list = ['critical_habitat_esa_listed_species.shp']
# ecological resources
process_shp(input_dir=eco_dir, output_dir=technology_specific_dir, crs=crs, skip_list=skip_list)
```

```python
# raster files
raster_list = ['ap_falcon_d.tif', 'b_eagle_d.tif', 'ca_condor_d.tif', 'g_eagle_d.tif', 'm_s_owl_d.tif', 'n_s_owl_d.tif']

for raster_file in raster_list:
    process_raster(input_raster_file=raster_file, input_dir= source_dir, output_dir=technology_specific_dir) 
```

## Recreation BLM Areas

```python
# recreation areas
process_shp(input_dir=rec_dir, output_dir=technology_specific_dir, crs=crs)
```

## Special Land Use Areas

```python
skip_list = ['national_monument.shp', 'national_natural_landmark.shp', 'nlcs_other.shp', 
             'wild_scenic_river.shp', 'wilderness_area.shp', 'wilderness_study_area.shp', 'acec.shp']

# special land use
process_shp(input_dir=special_dir, output_dir=technology_specific_dir, crs=crs, skip_list=skip_list)
```

## Incompatible Land Usage

```python
skip_list = ['blm_designated_utility_corridors.shp', 'drecp_variance_process_lands.shp', 
            'NPS_area_of_high_potential_conflict.shp']

# incompatible land use
process_shp(input_dir=incomp_dir, output_dir=technology_specific_dir, crs=crs, skip_list=skip_list)
```

## Visual Resources

```python
skip_list = ['national_scenic_trail.shp']

# visual resources
process_shp(input_dir=vis_dir, output_dir=technology_specific_dir, crs=crs, skip_list=skip_list)
```
